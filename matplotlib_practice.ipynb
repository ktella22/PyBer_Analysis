```python
# Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month 
x_axis = ["Jan", "Feb", "Mar", "Apr", "May", "June", "July", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulation
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Creat the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt( )
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create a plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston')

# Create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker = "*", color= "blue", linewidth=2, label="Boston")

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create the plot with object-oriented Interface
fig, ax= plt.subplots()
ax.plot(x_axis, y_axis, marker = "D", color= "green", linewidth=2, label = "Boston")

# Creat labels for axes
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")

# Create a title
ax.set_title("PyBer Fare by Month")

# Add y-axis limit
ax.set_ylim(0,45)

# Add grid
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# using plt.plot() to make scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# using scatter function
plt.scatter(x_axis, y_axis)
```

```python
# Annotate
plt.scatter(y_axis, x_axis, c='red', label = "Chicago")
# Add legend
plt.legend()
# Add title and axes labels
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
# limit x label
plt.xlim(0,50)
# invert y-axis
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis,y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(3*data)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# List comprehension
#new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
# OOI method
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
# Bubble chart
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Annotate the bubble chart
fig, ax = plt.subplots(figsize= (7,5))
ax.scatter(y_axis, x_axis, c= "xkcd:sky blue", alpha = 0.8, edgecolors ="black", lw=2, s= [ i*10 for i in y_axis], label= "Boston")
# Add edge color black
#ax.set_edge("black")
# Add legend
ax.legend()
# Add title
ax.set_title("PyBer Fare by Month")
# axes
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
# limite to x-axis
ax.set_xlim(0,50)
# Invert y-axis
ax.invert_yaxis()
```

```python
# Pie chart
plt.pie(y_axis, labels=x_axis)
plt.show
```

```python
explode_values = (0,0,0,0,0,0,0.3,0,0,0,0,0)
plt.subplots(figsize=(8,8));
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%');
```

```python
# Assign 12 colros, one for each month:
colors =["slateblue", "magenta", "lightblue","green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize = (8,8))
plt.pie(y_axis, 
        explode = explode_values,
        colors = colors,
        labels=x_axis,
       autopct="%.1f%%")

plt.show()
```

```python
# Using object-oriented interface
fig, ax = plt.subplots(figsize=(8,8))
colors =["slateblue", "magenta", "lightblue","green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
       explode = explode_values,
       startangle=100, counterclock=False,
       colors = colors,
       labels = x_axis,
       shadow = True,
       autopct= "%.1f%%")
# Add decimal 
plt.show()
```
